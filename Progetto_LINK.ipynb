## 4. For each ranking, consider only the most recent data point. For each university, compute the maximum difference between the rankings (e.g. for Aarhus University the value is 122-73=49). Notice that some rankings are expressed as a range

La colonna *world rank* del dataset **Times** contiene il simbolo '-' in quanto il dato è espresso sotto forma di range. Per tale motivo creo una formula che determini il valore medio del range.

```python
def norm(value_rank):
    
    if len(value_rank) <= 3:
        return int(value_rank)
    
    if len(value_rank) == 7:
        pattern = re.compile('^(\d+)\-(\d+)')
        occurence = pattern.search (value_rank)
        
        if occurence:
                up = int(occurence.group(2))
                low = int(occurence.group(1))
                central_value = (up+low)/2
                return int(central_value)
```

```python
times['world_rank']=times['world_rank'].str.replace('=','')
```

Credo una nuova colonna del dataset denominata *number_wr* che riporta il risultato dell'applicazione del formula **norm**.

```python
times['number_wr']=times['world_rank'].apply(norm)
```

```python
times['number_wr'].dtype
```

Definisco una formula per estrarre la percentuale di maschi e calcolarne il **numero**. La applico (tramite una *apply*) alla colonna female_male_ratio.

```python
def male_ratio(ratio):
           
    pattern = re.compile('([0-9]+)$')
    occurrence = pattern.search(ratio['female_male_ratio'])
    
    if occurrence:
        fraction_male = int(occurrence.group(1))/100
        total_students = ratio['num_students']
        number_male = round(fraction_male * total_students,0)
        
        return number_male
```

```python
times_mostrecent_rank['n_male']=times_mostrecent_rank.apply(male_ratio,axis=1)
```

```python
times_mostrecent_rank['n_male'].head()
```

```python
times_mostrecent_rank['n_male'].dtype
```

Calcolo per differenza il **numero di femmine**.

```python
times_mostrecent_rank['n_female']=times_mostrecent_rank['num_students']-times_mostrecent_rank['n_male']
```
